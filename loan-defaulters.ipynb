# Import Python Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Import Datasets

```python
train_data = pd.read_csv("/kaggle/input/loan-prediction-based-on-customer-behavior/Training Data.csv")
```

# Check How Dataset Looks Like

```python
train_data.head()
```

```python
train_data.shape
```

# Convert datatype of selected fields

```python
train_data["Profession"]=pd.factorize(train_data.Profession)[0]
train_data["CITY"]=pd.factorize(train_data.CITY)[0]
train_data["STATE"]=pd.factorize(train_data.STATE)[0]
train_data["Married/Single"]=pd.factorize(train_data['Married/Single'])[0]
train_data["House_Ownership"]=pd.factorize(train_data.House_Ownership)[0]
train_data["Car_Ownership"]=pd.factorize(train_data.Car_Ownership)[0]
```

# Drop the dependent variable from the train dataset

```python
xtrain=train_data.drop("Risk_Flag",axis=1)
ytrain=train_data["Risk_Flag"]
```

# Train/Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(xtrain, ytrain, test_size=0.2)
```

# Train your model

```python
from sklearn.ensemble import RandomForestClassifier
RFClassifier= RandomForestClassifier(criterion='entropy', random_state=0)
RFClassifier.fit(X_train,y_train)
```

```python
from xgboost import XGBClassifier
help(XGBClassifier)
```

```python
my_model = XGBClassifier(n_estimators=500, use_label_encoder=False)
my_model.fit(X_train, y_train, 
             early_stopping_rounds=5, 
             eval_set=[(X_test, y_test)],
             verbose=False)
```

# Predict on test_data set

```python
y_pred_0= RFClassifier.predict(X_test)
y_pred = my_model.predict(X_test)
```

```python
print(my_model)
y_pred
```

```python
y_pred_0
```

# Evaluate based on ROC_AUC 

```python
from sklearn.metrics import roc_auc_score
```

```python
auc = roc_auc_score(y_test, y_pred_0)
print(auc)
```

```python
auc = roc_auc_score(y_test, y_pred)
print(auc)
```

```python
# Save test predictions to file
output = pd.DataFrame({'id': X_test.index,
                       'risk_flag': y_pred})
output.to_csv('submission.csv', index=False)
```
